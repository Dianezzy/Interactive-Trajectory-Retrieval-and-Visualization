```python
extracted_file_path = "C:\\WuYihong\\Data\\nba_movement_data\\data\\extracted_data"

records = set()
for movement in tqdm(movement_file_list):
    extracted_data = []
    period = -1
    # movement = "0021500002.json"
    game_id = movement.split(".")[0]
    movement_file = os.path.join(movement_file_path, movement)

    with open(movement_file, "r") as f:
        data = ujson.load(f)
    
    records = set()
    for event in data['events']:
        if len(event['moments']) < 1:
            continue
        
        for moment in event['moments']:
            if moment[2] is None:
                continue
            
            period = moment[0]

            game_clock = moment[2]
            shot_clock = moment[3]

            if (period, game_clock) not in records:
                records.add((period, game_clock))
                ball_position = moment[5][0][2:]
                players = [item[:4] for item in moment[5][1:]]

                extracted_data.append({'period': period, 'game_clock': game_clock, 'shot_clock': shot_clock, \
                                       'ball_position': ball_position, 'player_position': players})
    
    extracted_data = sorted(extracted_data, key=lambda x: int(x['period']) * 720 + float(x['game_clock']))
    with open(os.path.join(extracted_file_path, movement), "w") as f:
        ujson.dump(extracted_data, f, indent=4)

    # break

```

```python
extracted_file_path = "C:\\WuYihong\\Data\\nba_movement_data\\data\\extracted_data"
matched_file_path = "C:\\WuYihong\\Data\\nba_movement_data\\data\\matched_data"
game_splits_path = "game_info\\game_splits"
half_court_info_path = "game_info\\half_court_info.json"

with open(half_court_info_path, "r") as f:
    half_court_info = ujson.load(f)

for movement in tqdm(movement_file_list):
    period = -1
    # movement = "0021500002.json"
    game_id = movement.split(".")[0]
    extracted_file = os.path.join(extracted_file_path, movement)
    game_splits_file = os.path.join(game_splits_path, movement)

    with open(extracted_file, "r") as f:
        extracted_data = ujson.load(f)

    with open(game_splits_file, "r") as f:
        game_splits_data = ujson.load(f)
    
    for split in game_splits_data:
        period = split[0]
        start = split[1]
        end = split[2]
        offensive_team = split[3]
        result = split[4]
        index = split[5]
        split_movement = []

        if period <= 2:
            court = half_court_info[game_id]["first_half"][offensive_team]
        else:
            court = half_court_info[game_id]["second_half"][offensive_team]
        
        for item in extracted_data:
            if item['period'] == period and start >= item['game_clock'] >= max(end - 2, 0):
                if court[1] >= item['ball_position'][0] >= court[0]:
                    split_movement.append(item)
        
        # if len(split_movement) < 25:
        #     print("Wrong in", game_id, index)
        
        split_movement = sorted(split_movement, key= lambda x: x['game_clock'] * -1)

        matched_game_path = os.path.join(matched_file_path, game_id)
        if not os.path.exists(matched_game_path):
            os.makedirs(matched_game_path)
        with open(os.path.join(matched_game_path, "{}.json".format(index)), "w") as f:
            ujson.dump(split_movement, f, indent=4)

    # break
```
